```python
df = pd.read_csv('data/moves_summary.csv')
df
```

```python
walking = df[df['Group'] == 'walking']
walking['Steps'].plot()
```

```python
s = pd.Series(walking['Steps'].values, index=walking['Date'].values)
s.plot()
```

```python
print('Total Mean Steps: ', s.mean())
print('Mean Steps for January, 2016: ', s['1/1/16':'1/31/16'].mean())
print('Mean Steps for February, 2016: ', s['2/1/16':'2/28/16'].mean())
print('Mean Steps for March, 2016: ', s['3/1/16':'3/28/16'].mean())
```

```python
print('Histogram of All Steps Data:', s.hist())
```
