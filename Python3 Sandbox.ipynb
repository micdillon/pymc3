```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

%matplotlib inline
```

```python
X = stats.uniform(0,2)
Y = stats.uniform()
x = np.linspace(X.ppf(0.01), X.ppf(0.99), 100)

# plot the PDFs
fig, ax = plt.subplots(2,1)
ax[0].plot(x, X.pdf(x) + Y.pdf(x), 'r-', lw=5, alpha=0.6, label='uniform pdf')
ax[1].plot(x, X.pdf(x) - Y.pdf(x), 'b-', lw=5, alpha=0.6, label='uniform pdf')
```

```python
# scatter plot

X = stats.uniform(-1, 1)
Y = stats.uniform(-1,1)

Xs = X.rvs(size=1000)
Ys = Y.rvs(size=1000)

fig, ax = plt.subplots(1,1)
ax.scatter(Xs-Ys, Xs+Ys)
```
